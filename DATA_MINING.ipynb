# Label Encoding

```python
# Tahap Encoding, Transformasi data kategorik jadi numerik
df3 = df2

# Mengganti nilai-nilai dalam subset dengan data boolean
df3.iloc[:,2:9] = df3.iloc[:,2:9].notna()

#Mengubah data sebelumnya menjadi data integer
df3.iloc[:,2:9] = df3.iloc[:,2:9].astype('int')

#mengubah menjadi tipe data kategori dan mengganti nilai-nilai dalam kolom tersebut dengan kode kategori yang sesuai
df3['Q23'] = df3['Q23'].astype('category').cat.codes #data ordinal
df3['Q29'] = df3['Q29'].astype('category').cat.codes
df3['Q11'] = df3['Q11'].astype('category').cat.codes
df3['Q16'] = df3['Q16'].astype('category').cat.codes

#untuk mentransformasi sehingga memiliki rata-rata nol dan deviasi standar satu.
scaler = StandardScaler()

# untuk menghitung rata-rata dan deviasi standar dari setiap fitur dalam DataFrame
df3 = scaler.fit_transform(df3)
df3
```

```python
#grafik elbow method
cost = []
K = range(1,11)

# Loop untuk Menghitung Cost untuk Setiap Jumlah Kluster
for num_clusters in list(K):
    kmodes = KMeans(n_clusters=num_clusters, init = "random", n_init = 11, verbose=1)
    kmodes.fit_predict(df3)
    cost.append(kmodes.inertia_)

# Plotting Elbow Curve
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

Dari grafik Elbow Method dapat dilihat bahwa kita dapat menentukan titik siku atau titik optimal berada pada cluster 2, sehingga kita dapat mengambil titik pada cluster 0,1,2 untuk analisis selanjutnya

```python
#k-means
kmean = KMeans(n_clusters=3, init="random", n_init=11, verbose=1)
clusters = kmean.fit_predict(df3)
clusters
```
